```python
import numpy as np
from matplotlib import pyplot as plt
```

#SPIN HALL NANO OSCILLATOR DYNAMICS

### parameters

```python
gamma = np.float64(17.6*10**10)
alpha=np.float64(0.0005)
she=np.float64(0.07)
H=np.float64(1.89)
t=np.float64(2*10**(-9))
j=np.float64(0.24*10**11)
e=np.float64(1.603*10**(-19))
h=np.float64(1.054*10**(-34))
m=np.float64(1313*(10**3))
eta=np.float64((h*she*j)/(2*e*m*t))
```

```python
step=np.float64(10**(-13))
theta=np.zeros(10000,dtype=np.float64())
phi=np.zeros(10000,dtype=np.float64())
theta[0]=np.pi/4
```

```python
p1=gamma/(1+alpha**2)
p2=gamma*alpha/(1+alpha**2)
p3=gamma*eta/(1+alpha**2)
```

```python
for i in range(1,10000):
  theta[i]=theta[i-1]+step*(-p2*H*np.sin(theta[i-1])-p3*np.cos(theta[i-1])*np.sin(phi[i-1]))
  phi[i]=phi[i-1]+step*(p1*H-p3*np.cos(phi[i-1])/np.sin(theta[i-1]))
```

```python
time=np.arange(0,10000)*np.float64(10**(-13))
```

```python
plt.plot(time,np.sin(theta)*np.cos(phi))
```

```python
from numpy.fft import fft, ifft
```

```python
X=fft(np.sin(theta)*np.cos(phi))
n=np.arange(0,10000)
T=np.float64(10**(-9))
freq=n/T
plt.plot(np.abs(X),freq)
```
